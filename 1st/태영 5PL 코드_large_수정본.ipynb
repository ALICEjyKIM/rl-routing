```python
# -----------------------------
# M 계산
# -----------------------------
M = {}
for i in sellers:
    for j in buyers:
        for k in transporters:
            for t in T:
                cap_k_t = transporters[k]["cap"][t]
                M[(i,j,k,t)] = min(sellers[i]["supply"], buyers[j]["demand"], cap_k_t)

# -----------------------------
# valid_tuple 생성
# -----------------------------
valid_tuple = []
for (i,j,k,t), mval in M.items():
    net_unit = buyers[j]["price"][t-1] - sellers[i]["price"][t-1] - transporters[k]["cost"]
    if net_unit >= 0 and (buyers[j]["price"][t-1] > 0) and (sellers[i]["price"][t-1] > 0) and (transporters[k]["cap"][t] > 0):
        valid_tuple.append((int(i), int(j), int(k), int(t)))
valid_tuple = list(set(valid_tuple))
valid_tuple = sorted(valid_tuple, key=lambda x: x[3])
```

```python
# -----------------------------
# 모델 생성
# -----------------------------
import gurobipy as gp
from gurobipy import GRB

model = gp.Model("5PL")

# -----------------------------
# 변수 정의
# -----------------------------
# 기본 변수
L = model.addVars(T, lb=0.0, name="L")
y = model.addVars(valid_tuple, lb=0.0, name="y")
z = model.addVars(valid_tuple, vtype=GRB.BINARY, name="z")

# MA용
p = model.addVars(T, lb=-GRB.INFINITY, name="p")
p_est = model.addVars(T, lb=-GRB.INFINITY, name="p_est")
E = model.addVars(T, lb=0.0, name="E")
B = model.addVars(T, vtype=GRB.BINARY, name="B")
M_val = 1e6

# -----------------------------
# 목적함수
# -----------------------------
# E: MA 추정치와 실제값 차이 
# L: 거래 참여자 간 최소 이윤
obj = gp.quicksum((buyers[j]["price"][t-1] - sellers[i]["price"][t-1] - transporters[k]["cost"]) * y[i,j,k,t]
                  for (i,j,k,t) in valid_tuple) - 0.01 * gp.quicksum(E[t] for t in T) + 0.01 * gp.quicksum(L[t] for t in T)
model.setObjective(obj, GRB.MAXIMIZE)

# -----------------------------
# 제약식
# -----------------------------
for (i,j,k,t) in valid_tuple:
    mval = M[(i,j,k,t)]

    # 균형 가격 제약 (거래 활성화시 이윤 보장)
    model.addConstr(p[t] >= sellers[i]["price"][t-1] - mval*(1 - z[i,j,k,t]), name=f"price_low_{i}_{j}_{k}_{t}")
    model.addConstr(p[t] + transporters[k]["cost"] <= buyers[j]["price"][t-1] + mval*(1 - z[i,j,k,t]), name=f"price_high_{i}_{j}_{k}_{t}")
    
    # L[t] 제약 (최소 이윤)
    model.addConstr(L[t] <= buyers[j]["price"][t-1] - p[t] - transporters[k]["cost"] + mval*(1 - z[i,j,k,t]), name=f"L_buyer_{i}_{j}_{k}_{t}")
    model.addConstr(L[t] <= p[t] - sellers[i]["price"][t-1] + mval*(1 - z[i,j,k,t]), name=f"L_seller_{i}_{j}_{k}_{t}")

# z=0 (거래X) 시점에선 L[t] = 0, p[t] = seller, buyer price 평균
for t in T:
    model.addConstr(L[t] <= M_val * gp.quicksum(z[i,j,k,t] for (i,j,k,tt) in valid_tuple if tt==t))
    sum_z = gp.quicksum(z[i, j, k, t] for (i, j, k, tt) in valid_tuple if tt == t)
    buyer_vals = [buyers[j]["price"][t-1] - transporters[k]["cost"]
                  for (i, j, k, tt) in valid_tuple if tt == t]
    seller_vals = [sellers[i]["price"][t-1]
                   for (i, j, k, tt) in valid_tuple if tt == t]
    if buyer_vals and seller_vals:
        p_mid = 0.5 * (sum(buyer_vals)/len(buyer_vals) + sum(seller_vals)/len(seller_vals))
    else:
        p_mid = p[t-1] if t > 0 else 0  # 첫 시점이면 초기값 0

    model.addConstr(p[t] >= p_mid - M_val * sum_z, name=f"p_lower_{t}")
    model.addConstr(p[t] <= p_mid + M_val * sum_z, name=f"p_upper_{t}")
    if t > 1:
        model.addConstr(p[t] <= p[t-1] + M_val * sum_z, name=f"p_upper_prev_{t}")
        model.addConstr(p[t] >= p[t-1] - M_val * sum_z, name=f"p_lower_prev_{t}")



# y <= M * z
for (i,j,k,t) in valid_tuple:
    model.addConstr(y[i,j,k,t] <= M[(i,j,k,t)] * z[i,j,k,t], name=f"act_{i}_{j}_{k}_{t}")

# seller supply
for i in sellers:
    model.addConstr(gp.quicksum(y[ii,j,k,t] for (ii,j,k,t) in valid_tuple if ii==i) <= sellers[i]["supply"],
                    name=f"supply_{i}")

# buyer demand
for j in buyers:
    model.addConstr(gp.quicksum(y[i,jj,k,t] for (i,jj,k,t) in valid_tuple if jj==j) <= buyers[j]["demand"],
                    name=f"demand_{j}")

# transporter capacity
for k in transporters:
    for t in T:
        model.addConstr(gp.quicksum(y[i,j,kk,tt] for (i,j,kk,tt) in valid_tuple if kk==k and tt==t)
                        <= transporters[k]["cap"][t], name=f"cap_{k}_{t}")

# 라이더는 한 시점당 한번만 운송 가능
for k in transporters:
    for t in T:
        # z[i,j,k,t]의 합 ≤ 1
        model.addConstr(
            gp.quicksum(z[i,j,k,t] for (i,j,kk,tt) in valid_tuple if kk==k and tt==t) <= 1,
            name=f"one_trip_per_time_{k}_{t}"
        )

# -----------------------------
# MA constraints (이동평균)
# -----------------------------
T_list = list(T)
alpha = 0.3
for idx, t in enumerate(T_list):
    if idx < 2:
        model.addConstr(p_est[t] == p[t], name="EMA_init")
    else:
        model.addConstr(p_est[t] == alpha * p[t-1] + (1-alpha) * p_est[t-1])

# 0.9*p_est[t] <= p[t] <= 1.1*p_est[t]일시 B[t] = 0 -> E[t] = 0 -> 목적함수 영향 X
# 범위 밖일시 E[t] >= |p[t]-p_est[t]| -> E[t] 최소화 위해 p_est와 가까워지도록 p 조정됨
for t in T:
    model.addConstr(p[t] - 0.9*p_est[t] >= -M_val*(B[t]))
    model.addConstr(1.1*p_est[t] - p[t] >= -M_val*(B[t]))
    
    model.addConstr(E[t] <= M_val*B[t])
    model.addConstr(E[t] >= p[t] - p_est[t] - M_val*(1 - B[t]))
    model.addConstr(E[t] >= -(p[t] - p_est[t]) - M_val*(1 - B[t]))


model.Params.InfUnbdInfo = 1

# Solve
# -----------------------------
model.setParam('OutputFlag', 1)
model.setParam('TimeLimit', 3600)
model.optimize()
```
